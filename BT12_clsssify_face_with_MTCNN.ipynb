```python
import matplotlib.pyplot as plt
from mtcnn import MTCNN
filename = '/content/download (6).jfif'
img = plt.imread(filename)
detector = MTCNN()
faces = detector.detect_faces(img)
plt.imshow(img)
ax = plt.gca()
for face in faces:
  x, y , width, height = face['box']
  vect = plt.Rectangle((x, y), width, height, fill = False, color = 'red') #fill = False kh tô hình chử nhật
  ax.add_patch(vect)  # để chồng hình chử nhật lên hình 
  for key,value in face ['keypoints'].items():
    dot = plt.Circle(value, radius = 2, color = 'red')
    ax.add_patch(dot)
plt.show()

```

```python
import matplotlib.pyplot as plt
from mtcnn import MTCNN
def draw_faces (filename, result_list):
  data = plt.imread(filename)
  for i in range (len(result_list)):
    x1, y1, width, height = result_list[i]['box']
    x2 = x1 + width
    y2 = y1 + width
    #plt.subplot(1, len(result_list), i + 1)
    plt.axis('off')
    plt.imshow(data[y1:y2, x1:x2])
    pos = '/content/drive/MyDrive/Data_CNN/face/' + str(i) + '.jpg'
    plt.savefig(pos)
  #plt.show()

filename = '/content/download (6).jfif'
img = plt.imread(filename)
detector = MTCNN()
faces = detector.detect_faces(img)
draw_faces(filename, faces)
```
