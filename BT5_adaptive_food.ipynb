```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
taste  = ctrl.Antecedent(np.arange(0,3,0.1),'taste')
environment = ctrl.Antecedent(np.arange(0,3,0.1),'environment')
budget = ctrl.Antecedent(np.arange(0,3,0.1),"budget")
userreview = ctrl.Antecedent(np.arange(0,3,0.1),'userreview')
location = ctrl.Antecedent(np.arange(0,3,0.1),"location")
range = ctrl.Consequent(np.arange(0,9,0.1),"range")

taste["Low"]=fuzz.gaussmf(taste.universe, 0, 1)
taste["Medium"]=fuzz.gaussmf(taste.universe, 1.5, 0.5)
taste["High"]=fuzz.gaussmf(taste.universe, 3, 1)
taste.view()

environment["Low"]=fuzz.gaussmf(environment.universe, 0, 1)
environment["Medium"]=fuzz.gaussmf(environment.universe, 1.5, 0.5)
environment["High"]=fuzz.gaussmf(environment.universe, 3, 1)
environment.view()

budget["Low"]=fuzz.gaussmf(budget.universe, 0, 1)
budget["Medium"]=fuzz.gaussmf(budget.universe, 1.5, 0.5)
budget["High"]=fuzz.gaussmf(budget.universe, 3, 1)
budget.view()

userreview["Low"]=fuzz.gaussmf(userreview.universe, 0, 1)
userreview["Medium"]=fuzz.gaussmf(userreview.universe, 1.5, 0.5)
userreview["High"]=fuzz.gaussmf(userreview.universe, 3, 1)
userreview.view()

location["Low"]=fuzz.gaussmf(location.universe, 0, 1)
location["Medium"]=fuzz.gaussmf(location.universe, 1.5, 0.5)
location["High"]=fuzz.gaussmf(location.universe, 3, 1)
location.view()

range["Level1"]=fuzz.trimf(range.universe,[0,1.5,3])
range["Level2"]=fuzz.trimf(range.universe,[1.5,3,4.5])
range["Level3"]=fuzz.trimf(range.universe,[3,4.5,6])
range["Level4"]=fuzz.trimf(range.universe,[4.5,6,7.5])
range["Level5"]=fuzz.trimf(range.universe,[6,7.5,9])
range.view()
```
